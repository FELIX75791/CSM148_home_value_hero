# Relationships between SalePrice and Numerical Variables

We will analyze some numerical variables which we think are important and explore their relationship to `SalePrice`.

```python
#scatter plot grlivarea/saleprice
var = 'GrLivArea'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

Above ground living area square feet (`GrLivArea`) is a important feature for predicting `SalePrice` since larger living area will generally result in higher price. The scatterplot confirms our assumption, showing a positive linear relationship between `GrLivArea` and `SalePrice`.

```python
#scatter plot totalbsmtsf/saleprice
var = 'TotalBsmtSF'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

Total square feet of basement area (`TotalBsmtSF`) has a strong positive linear relationship to `SalePrice`.

```python
var = 'BedroomAbvGr'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

We assume with more bedrooms above ground, the price will be higher. But from the scatterplot, we can see that the mean price doesn't increase as the number of bedroom increases. Therefore, we cannot conclude that they have a positive linear relationship.

# Identify Multicollinearity in Numerical Variables

```python
plt.figure(figsize=(30,30))
sns.heatmap(df_train.select_dtypes('number').corr(),vmax=.8, annot=True,square=True,cmap='viridis');
```

From the heatmap, we could see 'TotalBsmtSF' and '1stFlrSF' (corr = 0.82), 'GarageCars' and 'GarageAreas' (corr = 0.88), and 'YearBuilt' and 'GarageYrBlt' (corr = 0.83) have very high correlation. Those correlation is so strong that it can indicate a situation of multicollinearity, which we need to remove one of them during the data preprocessing since they probably provide the same information.

```python
quantitative = [f for f in df_train.columns if df_train.dtypes[f] != 'object']
quantitative.remove('SalePrice')
quantitative.remove('Id')
spr = pd.DataFrame()
spr['feature'] = quantitative
spr['spearman_corr'] = [df_train[f].corr(df_train['SalePrice'], 'spearman') for f in quantitative]
spr = spr.sort_values('spearman_corr')
plt.figure(figsize=(6, 0.25*len(quantitative)))
sns.barplot(data=spr, y='feature', x='spearman_corr', orient='h');
```

We can also examine the Spearman correlation, since it picks up monotonic relationships (could be nonlinear). We can see some deterministic features that played an important role, such as OverallQual, GrLivArea, etc, which could be helpful in feature selection and providing us some insights on feature importance.

# Missing Data Analysis

```python
#missing data
total = df_train.isnull().sum().sort_values(ascending=False)
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

We'll consider that when more than 15% of the data is missing, we should delete the corresponding variable and pretend it never existed since imputing will make the prediction inaccurate. According to this, there is a set of variables (e.g. 'PoolQC', 'MiscFeature', 'Alley', etc.) that we should delete. But as we scan through the missing values of categorical features, some of them are having missing values simply means the observation doesn't exhibit the features. For example, "BsmtQual" and "BsmtCond" exhibit NA values because some datapoint doesn't have a basement, but the features itself still provide meaningful information for rest of the datapoints. Therefore, we will identify those features with meaningful NA and create a new category "Not Applicable" to indicate they doesn't exhibit the property during data preprocessing.
