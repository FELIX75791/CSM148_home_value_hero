# Neural Network Model Analysis on House Prices

## Load the Preprocessed Data

```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the preprocessed training and testing data
X_train = pd.read_csv('./data/X_train.csv')
y_train = pd.read_csv('./data/y_train.csv').squeeze()  # Convert to a Series if not already
X_test = pd.read_csv('./data/X_test.csv')
y_test = pd.read_csv('./data/y_test.csv').squeeze()  # Convert to a Series if not already

# Initialize the Neural Network Model
# Here we choose a simple network with one hidden layer with 100 neurons as an example.
# This should be tuned according to the complexity of the dataset and the problem.
nn_model = MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam', 
                        alpha=0.0001, batch_size='auto', learning_rate='constant', 
                        learning_rate_init=0.001, max_iter=200, random_state=42)

# Train the model with the training data
nn_model.fit(X_train, y_train)

# Predict the Test set results
y_pred_nn = nn_model.predict(X_test)

# Evaluate the Neural Network Model Performance
mse_nn = mean_squared_error(y_test, y_pred_nn)
r2_nn = r2_score(y_test, y_pred_nn)

# Output model performance
print('Neural Network Model Performance:')
print('Mean Squared Error:', mse_nn)
print('R-squared Score:', r2_nn)

# Displaying neural network model settings
print('\nNeural Network Model Settings:')
print(nn_model.get_params())

# Plotting the true vs predicted values for visual analysis
plt.figure(figsize=(10,6))
plt.scatter(y_test, y_pred_nn)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
plt.xlabel('Measured')
plt.ylabel('Predicted')
plt.title('Neural Network Model - True Value vs Predicted Value')
plt.show()
```
